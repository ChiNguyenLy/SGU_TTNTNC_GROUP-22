```python
import os

def list_mazes(directory):
    txt_files = []
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            txt_files.append(filename)
    txt_files.sort()
    return txt_files

mazes = list_mazes(".")

for maze_file in mazes:
    with open(maze_file, "r") as f:
        maze_str = f.read()
    maze = mh.parse_maze(maze_str)
    
    print(maze_file)
    
    mh.show_maze(maze)
```
